```python
import random
import matplotlib.pyplot as plt
```

```python
x = []
y = []
size = 100

for i in range(size):
  x.append(random.randint(1, 500))
  y.append(random.randint(25, 70) * random.uniform(-1, 1) + x[i] * (x[i] // 10))
```

```python
line_x = [min(x), max(x)]
line_y = [min(y), max(y)]


w = (line_y[1] - line_y[0]) / (line_x[1] - line_x[0])
b =  line_y[0] - w * line_x[0]
pred_X = 87
pred_Y = w * pred_X + b


plt.plot(x, y, '+')
plt.plot(line_x, line_y, color='red')
plt.plot(pred_X, pred_Y, '*', color='green')
```
